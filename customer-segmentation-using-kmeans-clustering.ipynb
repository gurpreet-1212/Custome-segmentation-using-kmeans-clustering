```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv("/kaggle/input/malldata/Mall_Customers.csv")
```

```python
customer_data.head()
```

```python
customer_data.info()
```

```python
X=customer_data.iloc[:,[3,4]].values
```

```python
print(X)
```

#*****FINDING OF CLUSTERS

```python
import warnings
warnings.filterwarnings('ignore')
wcss=[]
for i in range(1,11):
    k_means = KMeans(n_clusters=i , init = "k-means++" , random_state = 42  )
    k_means.fit(X)
    
    wcss.append(k_means.inertia_)   
```

```python
sns.set()
plt.plot(range(1,11) , wcss)
plt.xlabel('Number_of_clusters')
plt.ylabel("WCSS")
plt.show()
```

# THE NUMBER OF CLUSTERS TO BE CHOSEN IS EQUAL TO 5 

```python
warnings.filterwarnings('ignore')
k_means1 = KMeans(n_clusters = 5 , init = "k-means++" , random_state= None )
Y= k_means1.fit_predict(X)
```

```python
print(Y)
```

```python
customer_data.shape
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[Y==0,0], X[Y==0,1] , s =50 , c="yellow" , label='Cluster_1')
plt.scatter(X[Y==1,0], X[Y==1,1] , s =50 , c="green" , label='Cluster_2')
plt.scatter(X[Y==2,0], X[Y==2,1] , s =50 , c="red" , label='Cluster_3')
plt.scatter(X[Y==3,0], X[Y==3,1] , s =50 , c="blue" , label='Cluster_4')
plt.scatter(X[Y==4,0], X[Y==4,1] , s =50 , c="orange" , label='Cluster_5')
plt.scatter(k_means1.cluster_centers_[:,0] , k_means1.cluster_centers_[:,1] , s=100 , c="black" , label="Centroid")
plt.xlabel('Annual Income')
plt.ylabel('Spending score')
plt.title('Clusters formed according to annual income and Spending score')
plt.show()
```
